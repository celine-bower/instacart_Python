```python
# more customers with 1 or more children than customers with no children.
# Exploring non-caregivers to see if they would purchase more in departments 'meat/seafood(12)- most expensive' and 'alcohol(5)'
# Explore spending in certain Departments with this subgroup:
# Incorporate age and income if needed (high_revenue_customer df also available):

Wealth = []

for value in df_merged['income']:
    if value > 100000:
        Wealth.append('upper')
    elif value >50000 and value <99999:
        Wealth.append('middle')
    else:
        Wealth.append('lower')
```

```python
df_merged['wealth'] = Wealth
```

```python
df_merged['wealth'].value_counts()
```

```python
# create flag for departments needed:
df_merged.loc[df_merged['department_id'] == 12, 'meat/seafood'] = 'meat_eater'
```

```python
df_merged.loc[df_merged['department_id'] != 12, 'meat/seafood'] = 'non_meat_eater'
```

```python
df_merged['meat/seafood'].value_counts()
```

```python
df_merged.loc[df_merged['department_id'] == 5, 'alcohol'] = 'alcohol_consumer'
```

```python
df_merged.loc[df_merged['department_id'] != 5, 'alcohol'] = 'non_alcohol_consumer'
```

```python
df_merged['alcohol'].value_counts()
```

```python
# splitting up age-groups (increase in income at age 40): 

age_group = []

for value in df_merged['Age']:
    if value < 20:
        age_group.append('under_20')
    elif value >=20 and value <30:
        age_group.append('20-29')
    elif value >=30 and value <40:
        age_group.append('30-39')
    elif value >=40 and value <50:
        age_group.append('40-49')
    elif value >=50 and value <60:
        age_group.append('50-59')
    else:
        age_group.append('over_60')
```

```python
df_merged['age_groups'] = age_group
```
